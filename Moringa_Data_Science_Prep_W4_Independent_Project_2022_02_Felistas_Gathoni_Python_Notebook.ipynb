# **1.0 LOADING THE DATASET.**


```python
# Import pandas
#
import pandas as pd
# Import numpy
#
import numpy as np
```

```python
# Load the dataset using this url (http://bit.ly/autolib_dataset)
#
Autolib_df = pd.read_csv('http://bit.ly/autolib_dataset')
Autolib_df

```

```python
# Accessing information about our dataset.
#
Autolib_df.info()
```

# **1.1 DATA CLEANING.**

```python
# Find missing data
# Count number of non-missing values
#
Autolib_df.count()
```

```python
# Count number of non-missing values
#
num_rows = Autolib_df.shape[0]
num_missing = num_rows - Autolib_df.count() 
num_missing
```

```python
# Since there are columns with Nan values only, we will drop them before removing rows with missing data and unwanted columns
# Drop irrelevant columns
#
Autolib_df.drop(["Cars", "Displayed comment", "Scheduled at", "Subscription status"], axis = 1, inplace = True) 
Autolib_df
```

```python
# Drop all rows with missing values
#

Autolib_clean_df = Autolib_df.dropna()
print(Autolib_clean_df)

# Verify there are no missing values
#
print(Autolib_clean_df.isnull().sum())
```

```python
# Drop all duplicate values
#
Autolib_df.drop_duplicates()
```

```python
# Check for outliers using a box plot
# Import seaborn library
import seaborn as sns

# Loading data from the city of Paris only
Autolib_paris = Autolib_df[Autolib_df['City']=='Paris'] 
bplot = sns.boxplot(y='hour', x='Bluecar counter', data=Autolib_paris, width=0.5, palette="colorblind")
bplot=sns.stripplot(y='hour', x='Bluecar counter',  data=Autolib_paris, jitter=True, marker='o', alpha=0.5, color='black')

bplot.axes.set_title("Paris: Hour Vs Bluecar counter", fontsize=14)
bplot.set_xlabel("Bluecar counter", fontsize=12)
bplot.set_ylabel("Hour", fontsize=12)
```

```python
# Drop outliers
#---
# Calculate the first quartile, third quartile and IQR

Q1 = Autolib_df.quantile(0.25)
Q3 = Autolib_df.quantile(0.75)
IQR = Q3 - Q1

# Remove Outliers

Autolib_df_out = Autolib_df[~((Autolib_df < (Q1 - 1.5 * IQR)) | (Autolib_df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(Autolib_df_out.shape)
```

# **1.2 DATA ANALYSIS**

```python
# make data relevant to bluecars, time and stations.
# create a new data set
Autolib_df = Autolib_df[["ID", "Bluecar counter", "City", "Public name", "year", "month", "day", "hour", "minute"]] 
Autolib_df

# Taking city to be paris only
AutolibParis_df = Autolib_df[Autolib_df["City"] == "Paris"]
AutolibParis_df

# create a new column Date/Time
AutolibParis_df['datetime'] = pd.to_datetime(AutolibParis_df[['year','month','day','hour','minute']])
AutolibParis_df.head()                                             


```

```python
# find number of cars returned
AutolibParis_df['Bluecar_used'] = AutolibParis_df['Bluecar counter'].diff()
AutolibParis_df


```

```python
# To find bluecars that were picked
#

AutolibParis_df['Bluecar_picked'] = AutolibParis_df[AutolibParis_df['Bluecar_used'] < 0].Bluecar_used.abs()
AutolibParis_df
```

```python
# Most popular hour of the day for picking a Bluecar in the city of Paris in April 2018
#
AutolibParis_df.groupby(['hour'])['Bluecar_picked'].agg('sum').sort_values(ascending=False)
```

The most popular hour is 9pm.

# **1.3 FURTHER ANALYSIS**

```python
# Most popular hour for returning cars
#
Autolib_df = pd.read_csv('http://bit.ly/autolib_dataset')
Autolib_df
#
Autolib_df.drop(["Cars", "Displayed comment", "Scheduled at", "Subscription status"], axis = 1, inplace = True) 
Autolib_df
#
Autolib_clean_df = Autolib_df.dropna()
print(Autolib_clean_df)
#
Autolib_df.drop_duplicates()
#
Q1 = Autolib_df.quantile(0.25)
Q3 = Autolib_df.quantile(0.75)
IQR = Q3 - Q1
```

```python
AutolibCars_df = Autolib_df[["ID", "Bluecar counter", "Utilib counter", "Utilib 1.4 counter", "City", "Postal code", "Public name", "year", "month", "day", "hour", "minute"]] 
AutolibCars_df
```

```python
AutolibCars_df['Cars'] = AutolibCars_df['Bluecar counter'] + AutolibCars_df['Utilib counter'] + AutolibCars_df['Utilib 1.4 counter']
AutolibCars_df                
```

```python
AutolibCars_df['Cars_used'] = AutolibCars_df['Cars'].diff()
AutolibCars_df
```

```python
AutolibCars_df['Cars_returned'] = AutolibCars_df[AutolibCars_df['Cars_used'] > 0].Cars_used.abs()
AutolibCars_df
```

```python
AutolibCars_df.groupby(['hour'])['Cars_returned'].agg('sum').sort_values(ascending=False)
```

6.00am is the most popular hour.

```python
# Most popular station overall?
#
AutolibCars_df.groupby(['ID'])['Cars'].agg('sum').sort_values(ascending=False)
```

```python
# Most popular station during the most popular hour
#
AutolibCars_df.groupby(['hour'])['Cars_used'].agg('sum').sort_values(ascending=False)
#
#AutolibCars_df.groupby(['ID'])[['hour'] == 6].agg('sum').sort_values(ascending=False)
AutolibCars_df = AutolibCars_df.loc[AutolibCars_df['hour'] == 6]
AutolibCars_df['ID'].sort_values(ascending=False)
```

```python
# Most popular postal code for Blue cars in overall
#
AutolibCars_df.groupby(['Postal code'])['Bluecar counter'].agg('sum').sort_values(ascending=False)
```

```python
# Most popular postal code for at the most popular picking hour
#
AutolibCars_df['Cars_picked'] = AutolibCars_df[AutolibCars_df['Cars_used'] < 0].Cars_used.abs()
AutolibCars_df
#
AutolibCars_df.groupby(['hour'])['Cars_picked'].agg('sum').sort_values(ascending=False)
#
AutolibCars_df = AutolibCars_df.loc[AutolibCars_df['hour'] == 6]
AutolibCars_df['Postal code'].sort_values(ascending=False)
```
